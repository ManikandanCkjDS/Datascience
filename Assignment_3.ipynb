1.  How many global scopes are there in a Python program? How many local
    scopes?

**Ans :**  One global scope, and a local scope is created whenever a
function is called.

1.  What happens to variables in a local scope when the function call
    returns?

**Ans :** Function returns, the local scope is destroyed, and all the
variables in it are forgotten.

1.  What is the concept of a return value? Is it possible to have a
    return value in an expression?

> **Ans :** Value that a function call evaluates to. Like any value, a
> return value can be used as part of an expression.
