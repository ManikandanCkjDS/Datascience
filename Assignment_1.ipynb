4\. What makes up an expression? What are the functions of all
expressions?

**<span class="underline">Expression</span>** – Combination Values and
Operators, Constant and variables.

**<span class="underline">Function of Expression :</span>**

1.  <span class="underline">**Constant expressions -** Only Constant
    Values. **Ex :** 1 , 1+2 / 2.0, ‘a’.</span>

2.  <span class="underline">**Integral expressions:** produce integer
    results after implementing all the automatic and explicit type
    conversions. **Ex :** a, a\*a, a + int(2.0).</span>

3.  <span class="underline">**Floating expressions**: produce floating
    point results after implementing all the automatic and explicit type
    conversions. **Ex**: a + b, 2.23.</span>

4.  <span class="underline">**Relational expressions:** Relational
    Expressions results of type bool which takes a value true or false.
    **Ex:** a \<= b, a + b \> 2.</span>

5.  <span class="underline">**Logical expressions:** Logical Expressions
    combine two or more relational expressions and produces bool type
    results. **Ex:** a \> b && a == 10, x == 10 \|\| b == 5.</span>

6.  <span class="underline">**Pointer expressions:** Pointer Expressions
    produce address values. **Ex:** &a ptr, ptr++.</span>

7.  <span class="underline">**Bitwise expressions:** Bitwise Expressions
    are used to manipulate data at bit level. They are basically used
    for testing or shifting bits. **Ex:** x \<\< 3 (shifts three bit
    position to left y) \>\> 1.</span>
